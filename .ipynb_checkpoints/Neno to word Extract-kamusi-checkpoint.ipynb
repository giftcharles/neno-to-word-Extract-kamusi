```python
import re
import pandas as pd
import random
```

```python
kamusi_file = "kamusi-swahili-english.txt"

with open(kamusi_file, "r") as f:
    maneno_list = f.read().replace("appl.", "application").split("\n")

maneno_list = [re.sub(r"-|ã€€", "", e) for e in maneno_list if e!="" and len(e) > 1]
len(maneno_list)
```

```python
random.choice(maneno_list)
```

```python
neno_lines=[]
for i in maneno_list:
    for l in i.split("."):
        if l!="":
            neno_lines.append(l.split(","))
            
neno_lines
```

```python
neno_word_trans_dictionary = []
for i in neno_lines:
    for l in i:
        swahili_word_regex = re.search(r"(\w+) (\(.*\))", l.strip())
        if swahili_word_regex != None and i.index(l) == 0:
            uwingi:str=""
            word:str=""
            if len(i) == 2:
                word = i[1].strip()
            elif len(i) == 3:
                if re.search(r"pl (\w+)", i[1].strip()) != None:
                    uwingi = re.search("pl (\w+)", i[1]).group(1)
                    
                word = re.sub(r"(\w+) [\(\w+\)]", "\1", i[2].strip())
            
            neno_word_trans_dictionary.append({"neno": swahili_word_regex.group(1),
                                               "tag": re.sub("[()]|\d/\d|\d+an|an|\d+|\s^|\n^", "", swahili_word_regex.group(2)),
                                              "word": word,
                                              "uwingi": uwingi})
            continue
neno_word_trans_dictionary     
```

```python
df = pd.DataFrame(data=neno_word_trans_dictionary)
df
```

```python
df.to_csv(f"neno-to-word-{df.shape[0]}.csv")
df.to_pickle(f"neno-to-word-{df.shape[0]}.pkl")
```
